```python
import pandas as pd
```

```python
categorical = pd.read_csv('categorical.csv')
categorical.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# loop through all categorical columns
for column in categorical.columns:
    # create a count plot
    sns.countplot(x=column, data=categorical)
    plt.show()
```

# I am dropping the customer column because of so many unique values and also it does not help much with data.

```python
categorical = categorical.drop(['customer'], axis=1)
```

```python
categorical.head()
```

```python
customer_df= pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
pd.set_option('display.max_columns', None)
customer_df.head()
```

```python
customer_df = customer_df.rename(columns={"Effective To Date": "effective_to_date"})
customer_df.head()
```

```python
numerical_df= pd.read_csv('numerical.csv')
numerical_df.head()
```

```python
numerical_df['month_year'] = pd.DatetimeIndex(customer_df['effective_to_date']).month
numerical_df['day_year'] = pd.DatetimeIndex(customer_df['effective_to_date']).day
numerical_df['year_year']= pd.DatetimeIndex(customer_df['effective_to_date']).year
numerical_df.head()
```
